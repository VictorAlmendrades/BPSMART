```python
import pandas as pd
import numpy as np
import requests
df=pd.read_json('https://www.datos.gov.co/resource/33dq-ab5a.json', typ='frame')
print(df)
```

```python

df = df.apply(lambda x: x.str.lower() if x.dtype == "object" else x) 
new = df["nme_convocatoria"].str.split(" ", n = 1, expand = True)

df['ano_convo'] = df['ano_convo'].str.replace("\W", "")
df['fcreacion_pd'] = df['fcreacion_pd'].str.replace("\W", "")

df['ano_convo'] = df['ano_convo'].str.replace("t", "")
df['fcreacion_pd'] = df['fcreacion_pd'].str.replace("t", "")
 
# making separate first name column from new data frame
df["Borrar"]= new[0]
 
# making separate last name column from new data frame
df["Info"]= new[1]
df.drop(columns =["nme_convocatoria"], inplace = True)
df.drop(columns =["Borrar"], inplace = True)

new = df["Info"].str.split(" ", n = 1, expand = True)
 
# making separate first name column from new data frame
df["nro_convocatoria"]= new[0]
 
# making separate last name column from new data frame
df["ano_convocatoria"]= new[1].str[-4:]

def whitespace_remover(dataframe):
   
    for i in dataframe.columns:
         
        if dataframe[i].dtype == 'object':
             
            dataframe[i] = dataframe[i].map(str.strip)
        else:
             
            pass
 
whitespace_remover(df)

df.dropna(inplace = True)
 
# new data frame with split value columns
new = df["id_producto_pd"].str.split("\W", n = 1, expand = True)
 
# making separate first name column from new data frame
df["First"]= new[0]
 
# making separate last name column from new data frame
df["Cod"]= new[1]
df.drop(columns =["id_producto_pd"], inplace = True)


new = df["Cod"].str.split("\W", n = 1, expand = True)
df["Cod"]= new[0]
df["Last"]= new[1]
df.drop(columns =["Cod"], inplace = True)
df.drop(columns =["Info"], inplace = True)
# Dropping old Name columns

 
# df display
print(df)



print(df)



print(df)
df.info(memory_usage='deep')
```

```python
df.to_csv('pandas.csv', sep=';')
```
